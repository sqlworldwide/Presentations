# SQL Assessment - Microsoft's Best Practices Checker

## Abstract

<span style="font-size: 14px;">Do you want to get the current 313 suggested best practices from the SQL Server Team and evaluate your environments? These best practices are available for all instances, from your Data Center to the cloud.</span>

In this demo session (Yes, no slides), I will show you both the “what” and “how” to use the SQL assessment API. How do you customize your checks and create rules that fit your environments? You will learn to identify a setting change after your initial review. The demo will use SQL assessment API, PowerShell cmdlets, and the new ‘SQL Assessment’ extension in Azure Data Studio to export assessment results in a SQL table and create an HTML report. I will show you how to configure "SQL best practices assessment configuration" in the Azure portal for SQL Server running on Azure VM.  
  

At the end of this session, you will be equipped with the toolset to set up and run your assessment, create customized rules, and determine your compliance score against suggested best practices.

## Why Would You Listen to This Talk?

- Is my SQL Server, SQL Server on Azure VM, Azure SQL Managed Instance configured with Microsoft recommended best practice?
    - Best practice rules are suggested by SQL Server Team.
    - Users can tune the default rules.
    - Create your own.
    - Bring rules created by third party. <mark>_(Do not use rulesets that come from untrusted sources until you thoroughly review them to make sure they are safe)_</mark>
    - Users can disable certain rules.

## Environments Supported

- SQL Server 2012 and higher
- SQL Server on Azure Virtual Machines
- <span style="font-size: 14px;">Azure SQL Database Managed Instance</span>
- <span style="font-size: 14px;">SQL on Linux-based systems</span>
- <span style="font-size: 14px;">SQL Server Assessment Extension for Azure Data Studio (ADS)</span>

## Warning!
<mark>The SQL Assessment API provides assessment on a variety of areas, but it does not go deeply into security. We recommend you use SQL Vulnerability Assessment to proactively improve your database security.</mark>
